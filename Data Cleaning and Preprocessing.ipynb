```python
import pandas as pd

df = pd.read_csv("Netflix_movies_and_tv_shows_clustering.csv")  # Replace with actual file
df
```

```python
# 1. Rename column headers to be lowercase and remove spaces
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# 2. Check and handle missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# 2. Handle missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unrated', inplace=True)
df['date_added'] = df['date_added'].apply(lambda x: pd.to_datetime(x, errors='coerce'))
```

```python
# 3. Remove duplicates
df.drop_duplicates(inplace=True)
df.dropna(subset=['date_added'], inplace=True)
```

```python
# 4. Standardize text formats
df['country'] = df['country'].str.strip().str.title()
df['type'] = df['type'].str.strip().str.title()
df['rating'] = df['rating'].str.strip().str.upper()
```

```python
# 5. Convert and split duration
df['duration'] = df['duration'].astype(str).str.strip()
df['duration_num'] = df['duration'].str.extract('(\d+)').astype(float)
df['duration_type'] = df['duration'].str.extract('([a-zA-Z]+)').fillna('Unknown')
```

```python
# 6. Final data type fix
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
```

```python
# 8. Summary
print(f"\n‚úîÔ∏è Cleaned dataset saved! Total rows: {len(df)}")
print(f"üîç Remaining missing values:\n{df.isnull().sum()}")
```

```python
# Step 7: Save cleaned dataset
df.to_csv("cleaned_netflix_data.csv", index=False)
```
